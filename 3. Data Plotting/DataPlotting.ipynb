1. Import Libraries:

**pandas** for data manipulation and analysis.
**matplotlib**.pyplot for data visualization.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

2. Data from each sheet is read into separate DataFrames.

```python
# Load the provided Excel file
CSV = "Cybersecurity_Datasets_Dummy.xlsx"

df_attack_types = pd.read_excel(CSV, sheet_name="Attack Types and Frequencies")
df_response_times = pd.read_excel(CSV, sheet_name="Incident Response Times")
df_monthly_incidents = pd.read_excel(CSV, sheet_name="Monthly Security Incidents")
df_budget_allocation = pd.read_excel(CSV, sheet_name="Security Budget Allocation")
```

3. **Scatter Plot**:

A scatter plot is created to *visualize the frequency of different attack types*.
The **x-axis** represents the *attack types*, and the **y-axis** represents their *frequencies*.

```python
# A Scatter Plot of Attack Types & their Frequencies
plt.figure(figsize=(10, 6))
plt.scatter(df_attack_types['Attack Type'], df_attack_types['Frequency'], color='b', label='Frequency')
plt.xlabel('Attack Type')  # Label - x-axis
plt.ylabel('Frequency')    # Label - y-axis
plt.title('Scatter Plot of Attack Types and Frequencies')  # Title
plt.xticks(rotation=45)    # Rotating x-axis labels (better readability)
plt.legend()               # Adding legend
plt.show()                 # Display
```

4. **Bar Chart**:

A bar chart is created to show the *frequency of different attack types*.


```python
# A Bar Chart for Monthly Incidents
plt.figure(figsize=(10, 6))
plt.bar(df_attack_types['Attack Type'], df_attack_types['Frequency'], color='g')
plt.xlabel('Attack Type')
plt.ylabel('Frequency')
plt.title('Bar Chart of Attack Types and Frequencies')
plt.xticks(rotation=45)
plt.show()
```

5. Bar Chart for **Monthly Incidents**:

A bar chart is created to *visualize the number of security incidents per month*.
The **x-axis** represents the *months*, and the **y-axis** represents the *number of incidents*.

```python
# Bar Chart for Monthly Incidents
df_monthly_incidents.plot(kind='bar', figsize=(10, 6), width=0.8)
plt.title('Monthly Security Incidents')
plt.xlabel('Month')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.show()
```

6. **Pie Chart**:

A pie chart is created to *show the distribution of the security budget* across different **categories**.

```python
# Pie Chart for Security Budget Allocation
plt.figure(figsize=(8, 8))
plt.pie(df_budget_allocation['Budget (USD)'], labels=df_budget_allocation['Category'], autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Security Budget Allocation')
plt.axis('equal')  # Ensuring that pie is drawn as a circle
plt.show()
```

7. **Subplots**:

- A 2x2 grid of subplots is created - **Scatter Plot**, **Bar Chart**, **Pie Chart**, and **Monthly Incidents Bar Chart** are added to individual subplots.

- **tight_layout()** is used to adjust the subplots to fit into the figure area.

```python
# Subplots

# Creating a figure with subplots (better correlation)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Scatter Plot
axes[0, 0].scatter(df_attack_types['Attack Type'], df_attack_types['Frequency'], color='b')
axes[0, 0].set_title('Scatter Plot of Attack Types and Frequencies')
axes[0, 0].set_xlabel('Attack Type')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].tick_params(axis='x', rotation=45)

# Bar Chart
axes[0, 1].bar(df_attack_types['Attack Type'], df_attack_types['Frequency'], color='g')
axes[0, 1].set_title('Bar Chart of Attack Types and Frequencies')
axes[0, 1].set_xlabel('Attack Type')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].tick_params(axis='x', rotation=45)

# Pie Chart
axes[1, 0].pie(df_budget_allocation['Budget (USD)'], labels=df_budget_allocation['Category'], autopct='%1.1f%%', startangle=140)
axes[1, 0].set_title('Pie Chart of Security Budget Allocation')
axes[1, 0].axis('equal')

# Bar Chart for Monthly Incidents
df_monthly_incidents.plot(kind='bar', ax=axes[1, 1], width=0.8)
axes[1, 1].set_title('Monthly Security Incidents')
axes[1, 1].set_xlabel('Month')
axes[1, 1].set_ylabel('Number of Incidents')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
